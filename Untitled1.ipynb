```python
# 1. Traemos a nuestra herramienta estrella
import pandas as pd

# 2. Leemos el archivo.
df = pd.read_csv('/content/credit_risk_dataset.csv')

# 3. Le pedimos que nos muestre las primeras 5 filas para verificar
df.head()
```

```python
# Comando 1: Nos da la radiografía del DataFrame (cuántas filas, columnas y si hay datos nulos)
df.info()
```

```python
# Comando 2: Nos da el resumen estadístico matemático (mínimos, máximos, promedios)
df.describe()
```

```python
# Filtramos a los mutantes conservando solo los datos lógicos
df = df[(df['person_age'] <= 80) & (df['person_emp_length'] <= 50)]

# Verificamos cuántas filas y columnas nos quedaron después de la purga
df.shape
```

```python
# Sumamos todos los valores nulos (vacíos) columna por columna
df.isnull().sum()
```

```python
# --- REGLA DE NEGOCIO: FILTRO DE ORIGINACIÓN ---
# Rechazo automático de registros con ingresos inconsistentes o
# por debajo del salario mínimo viable anualizado (USD $15,000)
# Esto previene que el modelo aprenda de datos anómalos o perfiles financieramente inviables.

clientes_iniciales = len(df)
df = df[df['person_income'] >= 15000]
clientes_filtrados = len(df)

print(f"Filtro de calidad de datos aplicado. Registros descartados: {clientes_iniciales - clientes_filtrados}")
```

```python
# 1. Rellenamos los vacíos usando la mediana de los clientes con la misma calificación de riesgo
df['loan_int_rate'] = df.groupby('loan_grade')['loan_int_rate'].transform(lambda x: x.fillna(x.median()))

# 2. Volvemos a pasar el radar para asegurarnos de que el problema desapareció
df.isnull().sum()
```

```python
# Convertimos todas las columnas de texto en banderas numéricas (0s y 1s)
# drop_first=True nos protege de la multicolinealidad matemática
df = pd.get_dummies(df, drop_first=True)

# Vemos cómo quedó nuestra nueva tabla numérica
df.head()
```

```python
# Traemos la herramienta para partir los datos
from sklearn.model_selection import train_test_split

# 1. Separamos las características (X) de nuestra variable objetivo (y)
X = df.drop('loan_status', axis=1) # X = Todas las columnas MENOS 'loan_status'
y = df['loan_status']              # y = SOLO la columna 'loan_status' (nuestra respuesta a predecir)

# 2. Dividimos los datos (80% para entrenar, 20% para probar)
# El random_state=42 asegura que si corremos esto mañana, haga exactamente el mismo corte aleatorio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# 3. Verificamos de qué tamaño quedó nuestro material de estudio y nuestro examen
print("Clientes para entrenar (80%):", X_train.shape)
print("Clientes para examinar (20%):", X_test.shape)
```

```python
# 1. Traemos la fórmula de la Regresión Logística
from sklearn.linear_model import LogisticRegression

# 2. Creamos nuestro modelo (le damos 'max_iter=1000' para darle suficiente tiempo de cálculo)
modelo_pd = LogisticRegression(max_iter=1000)

# 3. ¡A entrenar! Le pasamos las preguntas (X_train) y las respuestas correctas (y_train)
modelo_pd.fit(X_train, y_train)

print("¡El modelo ha terminado de estudiar y está listo para el examen!")
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Traemos al "traductor" que va a nivelar todas nuestras columnas
scaler = StandardScaler()

# 2. Le enseñamos a nivelar usando los datos de entrenamiento y los transformamos
X_train_scaled = scaler.fit_transform(X_train)

# 3. Transformamos el examen final (X_test) usando exactamente la misma escala
X_test_scaled = scaler.transform(X_test)

# 4. Volvemos a crear y entrenar el modelo, pero ahora con los datos limpios y escalados
modelo_pd = LogisticRegression(max_iter=1000)
modelo_pd.fit(X_train_scaled, y_train)

print("¡Modelo reentrenado con éxito y sin advertencias!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Le pedimos al modelo que resuelva el examen final (X_test_scaled)
predicciones = modelo_pd.predict(X_test_scaled)

# 2. Comparamos sus respuestas vs la realidad (y_test) para crear la matriz
matriz = confusion_matrix(y_test, predicciones)

# 3. Dibujamos el mapa de calor para el Director
plt.figure(figsize=(8,6))
sns.heatmap(matriz, annot=True, fmt='d', cmap='Reds')
plt.title('Matriz de Confusión del Motor Alpha')
plt.xlabel('Predicción del Modelo (0 = Paga, 1 = Default)')
plt.ylabel('Realidad (0 = Paga, 1 = Default)')
plt.show()

# 4. Imprimimos el reporte estadístico de precisión
print(classification_report(y_test, predicciones))
```

```python
# 1. Extraemos las probabilidades crudas (predict_proba)
# Esto nos devuelve dos columnas: [Probabilidad de Pagar (0), Probabilidad de Default (1)]
probabilidades = modelo_pd.predict_proba(X_test_scaled)

# 2. Solo nos interesa la columna 1 (el riesgo de Default), así que la separamos
probabilidad_default = probabilidades[:, 1]

# 3. Imprimimos los porcentajes de los primeros 5 clientes del examen
print("Probabilidades de Default (PD) de los primeros 5 clientes:")
print(probabilidad_default[:5])
```

```python
# 1. Creamos una nueva tabla especial para finanzas copiando el examen final original (sin estandarizar, para ver el dinero real)
tabla_finanzas = X_test.copy()

# 2. Le agregamos la columna de Probabilidad de Default (PD) que calculamos hace un momento
tabla_finanzas['PD'] = probabilidad_default

# 3. Definimos nuestro LGD bancario (Asumimos que perdemos el 80% del dinero si hay default)
LGD = 0.80

# 4. Calculamos la Pérdida Esperada (Expected Loss) en dinero real
# EAD es la exposición, es decir, el monto del préstamo original ('loan_amnt')
tabla_finanzas['Expected_Loss'] = tabla_finanzas['PD'] * LGD * tabla_finanzas['loan_amnt']

# 5. Ordenamos la tabla para ver a los clientes que más dinero nos harían perder y mostramos el Top 5
tabla_finanzas[['loan_amnt', 'PD', 'Expected_Loss']].sort_values(by='Expected_Loss', ascending=False).head()
```

```python
# --- REFRESCAMOS LA MEMORIA (Misión 14) ---
tabla_finanzas = X_test.copy()
tabla_finanzas['PD'] = probabilidad_default
LGD = 0.80

# --- APLICAMOS EL CRITERIO DE KELLY (Misión 15) ---
capital_banco = 2500000 # Nuestro Bankroll (Capital total del banco)

# 1. Definimos las probabilidades (p y q)
tabla_finanzas['p'] = 1 - tabla_finanzas['PD']  # Probabilidad de éxito
tabla_finanzas['q'] = tabla_finanzas['PD']      # Probabilidad de fracaso

# 2. Calculamos el ratio 'b' (Tasa de interés / Severidad de la pérdida)
# Nota: Dividimos la tasa entre 100 porque viene en formato 15.0 en lugar de 0.15
tabla_finanzas['b'] = (tabla_finanzas['loan_int_rate'] / 100) / LGD

# 3. Aplicamos tu fórmula sagrada: Criterio de Kelly
tabla_finanzas['Kelly_Fraction'] = tabla_finanzas['p'] - (tabla_finanzas['q'] / tabla_finanzas['b'])

# 4. Regla de Negocio: Si Kelly es negativo, asignamos 0. (Usamos una función lambda para limpiar)
tabla_finanzas['Kelly_Fraction'] = tabla_finanzas['Kelly_Fraction'].apply(lambda x: x if x > 0 else 0)

# 5. Calculamos el Límite de Crédito multiplicando la fracción por nuestro Bankroll
tabla_finanzas['Limite_Credito_MXN'] = tabla_finanzas['Kelly_Fraction'] * capital_banco

# 6. Ordenamos para ver a nuestros clientes Diamante (A los que más dinero les vamos a prestar)
columnas_finales = ['loan_amnt', 'loan_int_rate', 'PD', 'Kelly_Fraction', 'Limite_Credito_MXN']
tabla_finanzas[columnas_finales].sort_values(by='Limite_Credito_MXN', ascending=False).head()
```

```python
# 1. Aplicamos tu Fracción de Kelly (Un décimo: 0.10) para proteger el banco
fraccion_kelly = 0.10
tabla_finanzas['Kelly_Ajustado'] = tabla_finanzas['Kelly_Fraction'] * fraccion_kelly

# 2. Recalculamos el límite sugerido con la nueva fracción conservadora
tabla_finanzas['Limite_Sugerido'] = tabla_finanzas['Kelly_Ajustado'] * capital_banco

# 3. Aplicamos tu regla de oro: El tope máximo de $50,000 MXN
tope_maximo = 50000
# La función 'min' compara el límite sugerido vs los 50,000 y se queda siempre con el más bajo
tabla_finanzas['Limite_Final_MXN'] = tabla_finanzas['Limite_Sugerido'].apply(lambda x: min(x, tope_maximo))

# 4. Veamos cómo quedó nuestra cartera REAL lista para operar (Top 10 clientes)
columnas_finales = ['loan_amnt', 'PD', 'Limite_Sugerido', 'Limite_Final_MXN']
tabla_finanzas[columnas_finales].sort_values(by='Limite_Final_MXN', ascending=False).head(10)
```

```python
# 1. Calculamos la Capacidad de Pago: Máximo el 10% de su salario anual
tabla_finanzas['Cap_Ingresos'] = tabla_finanzas['person_income'] * 0.10

# 2. El duelo a muerte: Comparamos el Límite de Kelly vs el Límite de su Sueldo
# La función .min() agarrará automáticamente el número más bajo para protegernos
tabla_finanzas['Nuevo_Limite'] = tabla_finanzas[['Limite_Sugerido', 'Cap_Ingresos']].min(axis=1)

# 3. Aplicamos tu techo de cristal de $50,000 al nuevo límite ganador
tabla_finanzas['Limite_Final_MXN'] = tabla_finanzas['Nuevo_Limite'].apply(lambda x: min(x, 50000))

# 4. Veamos cómo quedó el promedio real del crédito asignado
print("Resumen de nuestra nueva cartera de créditos (MXN):")
print(tabla_finanzas['Limite_Final_MXN'].describe())

# 5. Volvemos a exportar el Excel ya corregido
tabla_finanzas.to_csv('cartera_motor_alpha_v2.csv', index=False)
print("\n¡Archivo actualizado! Descarga 'cartera_motor_alpha_v2.csv'.")
```
